## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Question:** Which pages should a content editor review first for refresh, expansion, protection, pruning, or monitoring?

**Unit of analysis:** one content page (one row = one page, `content_id`).

**Decision it improves:** an editor has limited time and cannot review every page. Right now they would have to pick pages by gut feeling. This work turns that into a ranked list.

**Who acts, and what do they do:** a content editor or SEO reviewer opens the top of the ranked queue (for example, the top 20-50 pages) and decides for each one whether to refresh, expand, protect, prune, or monitor it.

**Cost of a wrong recommendation:**
- False positive (page flagged but not actually a problem): wastes the editor's limited review time on a page that didn't need it.
- False negative (a truly declining page never makes the list): the page keeps losing traffic/visibility unnoticed, which is the more expensive mistake since it can go unfixed for a long time.

**Why data/ML can help here:** a simple hand-written rule (for example "flag if impressions are high and the page hasn't been updated in 6 months") only looks at one or two signals at a time. But there are many signals that matter together (traffic, position, freshness, CTR, engagement) and they interact in ways that are hard to write by hand. My own starter-pipeline run shows a learned model finds patterns a fixed rule misses (see numbers below), so this isn't a case where a plain rule is already good enough.


## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
from pathlib import Path

csv_candidates = [
    Path("data/raw/content_refresh_anonymized.csv"),
    Path("../../data/raw/content_refresh_anonymized.csv"),
]
csv_path = next(p for p in csv_candidates if p.exists())
df = pd.read_csv(csv_path)

total = len(df)
down = (df["trend_direction"] == "down").sum()
declining_with_demand = ((df["trend_direction"] == "down") & (df["impressions_90d"] >= 100)).sum()
low_ctr_visible = (
    (df["impressions_90d"] >= 500)
    & (df["avg_position"] > 0) & (df["avg_position"] <= 20)
    & (df["ctr"] < 0.5)
).sum()

print(f"Total pages: {total}")
print(f"Pages currently trending down: {down} ({down/total:.1%})")
print(f"Declining pages that still have real demand (impressions_90d >= 100): {declining_with_demand} ({declining_with_demand/total:.1%})")
print(f"Visible pages with weak CTR for their position (impressions>=500, position 1-20, CTR<0.5%): {low_ctr_visible} ({low_ctr_visible/total:.1%})")
```
